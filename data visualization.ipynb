# Relational plot

- when analysing with 2 num columns

-      1. line plot
-      2. scatter plot
-      3. rel plot

```python
data.columns
```

```python
# how to select the particular column from data
# data['col_name']
```

```python
# line plot

# sns.lineplot(x,y)

sns.lineplot(data = data, x ='sepal_length', y= 'sepal_width' )
```

```python
# sepal length vs sepal width with 3 species

sns.lineplot(x = data['sepal_length'], y = data['sepal_width'],hue = data['species'])
```

```python
sns.lineplot(x = data['sepal_length'], y = data['sepal_width'],hue = data['species'],ci = None)

# ci -- confidence intervel   -- population density
```

```python
# to ignore warning

import warnings
warnings.filterwarnings('ignore')
```

```python
# scatter plot

sns.scatterplot(x = data['sepal_length'], y = data['sepal_width'])
```

```python
sns.scatterplot(x = data['sepal_length'], y = data['sepal_width'],hue = data['species'])
```

```python
# rel plot
# by default its a scatter plot
# always we can convert rel plot into line plot by giving kind = line


sns.relplot(x = data['sepal_length'], y = data['sepal_width'])

sns.relplot(x = data['sepal_length'], y = data['sepal_width'], kind = 'line')
```

```python
sns.relplot(x = data['petal_length'], y = data['petal_width'],hue= data['species'])
```

# distribution plot

- analyse single num column
- histogram

```python
#histogram

sns.histplot(x=data['petal_length'],kde = True)
```

```python
# distplot
#its a combination of histplot and kde

sns.distplot(x= data['petal_length'])
```

# Categorical Plot

- used to analyze one numerical column and categorical column
- 1. count plot
  2. bar plot
  3. box plot
  4. violin plot
  5. strip
  6. swarm
  7. cat plot

```python
# 1. Count plot

# used to analyze the single categorical column
# gives count of variable
```

```python
sns.countplot(x = data['species'])
```

```python
# 2.box plot

# used to find the outliers

sns.boxplot(x=data['sepal_width'])
```

```python
sns.boxplot(x= data['species'], y= data['sepal_width'])
```

```python
# 3. bar plot

# used to analyze one num col and 1 cat col

sns.barplot(x = data['species'],y= data['petal_length'])
```

```python
# 4.violin plot

sns.violinplot(x = data['sepal_width'])
```

```python
#analyze

# the cenral box -- IQR, contains 50% of the data
# the thick line -- median
# the violin shape on either side of the box is kde of data distribution
```

```python
sns.violinplot(y = data['sepal_width'],x = data['species'])
```

```python
# 5. strip plot

# used to display individual data points

sns.stripplot(x=data['species'], y = data['sepal_width'],hue=data['species'])
```

```python
# 6. swarm plot

#type of scatter plot
#it shows all the individual datapoints and arranges to avoid the overlap of datapoints

sns.swarmplot(x=data['species'], y = data['sepal_width'],hue=data['species'])
```

```python
# 7. catplot

# using catplot we can plot box, violin, strip, swarm plot

sns.catplot(x=data['species'], y = data['sepal_width'],hue=data['species'])
# by default its a strip plot
```

```python
sns.catplot(data= data, x= 'species', y ='sepal_width', hue='species',kind='box')
```

```python
sns.catplot(data= data, x= 'species', y ='sepal_width', hue='species',kind ='violin')
```

```python
sns.catplot(data= data, x= 'species', y ='sepal_width', hue='species',kind='swarm')
```

```python
# Pair plot

sns.pairplot(data)

#sns.pairplot(data,kind = 'reg')
```

```python
# heatmap

data1 = data.drop('species',axis=1)   #dropped species- cat col

sns.heatmap(data1.corr(),annot=True,cmap="cubehelix")
```
